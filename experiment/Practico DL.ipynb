## Preparando los dataset

#### meli-challange-2019
Creamos la carpeta data en la raiz del proyecto, con los archivos de meli-challange-2019
No es necesario correrlo si ya existe el file

<code>%%bash

mkdir -p data
curl -L https://cs.famaf.unc.edu.ar/\~ccardellino/resources/diplodatos/meli-challenge-2019.tar.bz2 -o ./data/meli-challenge-2019.tar.bz2
tar jxvf ./data/meli-challenge-2019.tar.bz2 -C ./data/</code>


#### SBW-vectors
Dentro de la carpeta data anteriormente creada, subimos los files de SBW-vectors

<code>%%bash

curl -L http://cs.famaf.unc.edu.ar/\~ccardellino/SBWCE/SBW-vectors-300-min5.txt.bz2 -o ./data/SBW-vectors-300-min5.txt.bz2
bunzip2 ./data/SBW-vectors-300-min5.txt.bz2</code>
